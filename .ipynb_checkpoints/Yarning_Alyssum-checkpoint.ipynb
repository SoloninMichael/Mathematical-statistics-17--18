```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib
import statsmodels.api as sm 
from scipy.stats import rv_discrete
import scipy.special as sc
%matplotlib inline
```

```python
n = 30
Tn = []
P_lst = []
```

```python
class generate(rv_discrete):
    def _pmf(self, j = 1):
        return 6/(np.pi**2*j**2)
```

```python
one_x = generate(name="gen")
```

```python
def P(i):
    #print 6/(np.pi**2*i**2)
    return 6/(np.pi**2*i**2)
```

```python
def get_rv(n=1):
    u = np.random.random()
    j = 1
    tmblr = True
    while tmblr:
        sum_p_i = 0
        sum_p_i_plus_1 = 0
        for i in xrange(1, j+1):
            sum_p_i += P(i)
        sum_p_i_plus_1 = sum_p_i + P(j+1)
```

```python
def get_max_x(n = 1):
    tmp = []
    for i in xrange(1, n+1):
        tmp.append(one_x.pmf(i)) 
    #tmp.sort()
    #print "----", tmp, '-----', len(tmp)
    return max(tmp)
```

```python
def make_Tn ():
    for i in xrange(1, n+1):
        max_x = get_max_x(i)
        #print '###', max_x
        Tn.append(np.pi**2*max_x/(6*(i+1)))
```
