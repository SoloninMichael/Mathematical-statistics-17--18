```python
import numpy as np
import pandas as pd
import scipy
from scipy import optimize
from scipy.optimize import fmin
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.random(3)
```

```python
n = 1000
```

```python
Tn = []
ksi_n = []
etta_n = []
dzetta_n = []
X = np.random.random(n)
X.sort()
for i in xrange(n-1):
    num = int(math.ceil(np.sqrt(i+1)))
    tmp = (X[num] - (i+1)**(-1.0/2.0))*(i+1)**(3.0/4.0)
    Tn.append(tmp)
    ksi_n.append(X[i+1])
    etta_n.append(X[num])
    dzetta_n.append(X[num]*np.sqrt(i+1))
```

```python
p = plt.figure(figsize=(20,10))
p = plt.title('Tn sequence')
p = plt.plot(Tn, c = "tab:blue", marker = '.', ls = '-')

##какая здесь предельная функция ?????
t = np.arange(0.1, 10.5, 0.01) # 9.5
s = t*0+0.01
p = plt.plot(t, s, lw=3, c = "red")
plt.show()
```
